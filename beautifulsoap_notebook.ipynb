# 여러 페이지에서 모든 댓글 가져오기 

```python
import pandas as pd 
dataframe = pd.DataFrame(columns=['제목','평점','댓글'])
dataframe
```

```python
import time
import random

url = "https://movie.naver.com/movie/point/af/list.naver?&page="

for page_number in range(1,50):
    response = requests.get(url+str(page_number))
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        user_comment_list = soup.select('tbody > tr > td.title ')
        for i,td in enumerate(user_comment_list):
            movie_title = td.select_one('a.movie').text.strip()
            score = td.select_one('em').text.strip()
            comment = td.select_one('br').next_sibling.strip()
            dataframe.loc[i + 10*(page_number-1)] = [movie_title, score, comment]
        time.sleep(random.uniform(0.1,0.5))
        
dataframe
```

```python
dataframe.to_csv('평점정보.csv', sep=',', header=True, encoding='utf-8-sig')
```
