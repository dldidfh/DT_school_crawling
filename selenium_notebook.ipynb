# 테이블 헤더 파싱

```python

tabel_header = driver.find_elements_by_css_selector('thead > tr > th > span > span:nth-child(2)  ')
header_html = tabel_header[0].get_attribute('innerHTML')
print(header_html)
```

# 테이블 헤더 정보 저장 

```python
dataframe_header = []
for one_tabel_header in tabel_header:
    header = one_tabel_header.text
    dataframe_header.append(header)
    
print(len(dataframe_header))
dataframe_header
```

# 테이블 값 저장 

```python
dataframe_columns = []

row = []

for i, one_data in enumerate(find_all_tabel):
    if i % len(dataframe_header) == 0 and i !=0:
        dataframe_columns.append(row)
        row = []
    row.append(one_data.text)
    
    

print(len(row))
dataframe_columns[:5]
```

# dataframe 생성 

```python
import pandas as pd 
dataframe = pd.DataFrame(dataframe_columns, columns=dataframe_header)
dataframe
```

# 재사용 코드 함수화 

```python

def data_load(find_all_tabel):
    dataframe_columns = []
    row = []
    for i, one_data in enumerate(find_all_tabel):
        row.append(one_data.text)
        if len(row) % len(dataframe_header) == 0 and i !=0:
            dataframe_columns.append(row)
            row = []
    return dataframe_columns
    
```

# 모든 페이지 조회 

```python
buttons = driver.find_elements_by_css_selector('.pagination > li > a')

print([btn.text for btn in buttons])
# 다음과 같은 코드는 한문장으로 작성하는것이 더욱 효율적이다 
# list1 = []
# for btn in buttons :
#     list1.append(btn.text)
# print(list1)
```

```python
for btn in buttons:
    if btn.text == '›':
        print('존재합니다.')
    
```

# 페이지 조회하여 데이터 크롤링 

```python
data_frame_columns_list = []
q = 0
while True:
    q +=1
    if q > 3 : break   
    
    find_all_tabel = driver.find_elements_by_css_selector('tbody > tr > td')    
    nomalized_data = data_load(find_all_tabel)
    
    data_frame_columns_list = data_frame_columns_list + nomalized_data
    
    buttons = driver.find_elements_by_css_selector('.pagination > li > a ')
    
    if '›' in [btn.text for btn in buttons] and 'disabled' not in [btn.get_attribute('class') for btn in buttons]:
        [btn.click() for btn in buttons if btn.text == '›']
        time.sleep(0.5)
    else:
        break
    
        
    
```

# CSV 파일 저장 

```python
dataframe = pd.DataFrame(data_frame_columns_list, columns=dataframe_header)
dataframe
```

```python
dataframe.to_csv('created_data.csv', sep=',', header=True, encoding='utf-8-sig')
```
